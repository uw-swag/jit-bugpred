```python
import numpy as np
import pandas as pd
import os
from os import listdir
from os.path import isfile
import datetime

data_path = '../data/'
```

```python
tbal_df = pd.read_csv(data_path + 'balance_train.csv')
tbal_df
```

```python
tbal_df[tbal_df['buggy'] == True].groupby('strata').size().reset_index(name='count')
```

```python
sambt = tbal_df[tbal_df['buggy'] == True].sample(n=1125)
sambt
```

```python
sambt.groupby('strata').size().reset_index(name='count')
```

```python
tbal_df[tbal_df['buggy'] == False].groupby('strata').size().reset_index(name='count')
```

```python
samct = tbal_df[tbal_df['buggy'] == False].sample(n=6375)
samct
```

```python
samct.groupby('strata').size().reset_index(name='count')
```

```python
samt = pd.concat([sambt, samct]).sort_values('strata')
samt
```
